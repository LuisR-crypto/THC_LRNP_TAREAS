Al multiplicar la matriz por la transpuesta, lo que se genera es una matriz cuadrada, la cual va a poder ser invertible.


Todas las ecuaciones se pueden escribir como una multiplicación dematrices, las redes neuronales son matrices, las bases de datos también podrían considerarse.

La matriz que se genera de los coeficientes se le llama matriz de Vandermode, la matriz de la que hablo es la que en el esbozo es A

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA

def get_data(N=20):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1    #Esta es notación científica, significa 1 por 10 a la menos uno

  for i in range(N):
    y[i]=2*x[i]+3+(eps*(-1)**i)

  return x,y
```

```python
x,y=get_data(40)
plt.plot(x,y,"ro")
```

```python
x,y=get_data(40)
def plot_data(x,y):
  z=2*x+3
  plt.plot(x,y,"bo", label="Samples")
  plt.plot(x,z, "-m", label="Model")
  plt.title("Linear Regression")
  plt.xlabel("X data")
  plt.ylabel("Y data")
  plt.grid(lw=2)
  plt.legend()
  plt.show()
  x,y=get_data(40)
plot_data(x,y)
```

```python
x,y=get_data(40)
def plot_data(x,y, params):
  z=params[0]+params[1]*x

  plt.figure()
  plt.plot(x,y,"bo", label="Samples")
  plt.plot(x,z, "-m", label="Model")
  plt.title("Linear Regression")
  plt.xlabel("X data")
  plt.ylabel("Y data")
  plt.grid()
  plt.legend()
  plt.show()
```

```python
def linear_regression(N=20):
   x,y=get_data(N)
   A=np.vstack([np.ones(len(x)),x]).T
    #el b es la y
   AtA=A.T@A
   Aty=A.T@y
   params=LA.solve(AtA,Aty)
   plot_data(x,y,params)
   print(params)
   return params
```
