```python
x=np.linspace(0,20,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
#También existe la posibilidad de poner ambos juntos plt.plot(x,y1,x,y2)
plt.show()
```

```python
x=np.linspace(0,20,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1, marker="o")
plt.plot(x,y2, marker="*")
#De esdta forma se pone la froma de los indicadores en cada punto del gráfico y también se conectan con una línea
plt.show()
```

#Etiquetas

Siempre ponerle título a la gráfica, y a los ejes ponerles la información de lo que se les asoció

```python
x=np.linspace(0,20,20)
y1=x**2
plt.plot(x,y1, ls=":")

plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de la función x^2")
plt.show()

```

```python
x=np.linspace(0,20,20)
y1=x**2
plt.plot(x,y1, ls=":")

font1 = {'family':'serif','color':'m','size':20}
font2 = {'family':'serif','color':'c','size':15}

plt.xlabel("Eje X", fontdict=font2)
plt.ylabel("Eje Y", fontdict=font2)
plt.title("Gráfica de la función x^2", fontdict=font1)
plt.show()


```

```python
x=np.linspace(0,20,20)
y1=x**2
plt.plot(x,y1, ls=":")

font1 = {'family':'serif','color':'m','size':20}
font2 = {'family':'serif','color':'c','size':15}

plt.xlabel("Eje X", fontdict=font2)
plt.ylabel("Eje Y", fontdict=font2)
plt.title("Gráfica de la función x^2", fontdict=font1)
plt.grid()#Sirve para poner la retícula
plt.show()
```

```python
x=np.linspace(0,20,20)
y1=x**2
plt.plot(x,y1, ls=":")

font1 = {'family':'serif','color':'m','size':20}
font2 = {'family':'serif','color':'c','size':15}

plt.xlabel("Eje X", fontdict=font2)
plt.ylabel("Eje Y", fontdict=font2)
plt.title("Gráfica de la función x^2", fontdict=font1)
plt.grid(axis="x")#Sirve para formar la retícula solo de los úntos del eje x
plt.show()
```

```python
x=np.linspace(0,20,20)
y1=x**2
plt.plot(x,y1, ls=":")

font1 = {'family':'serif','color':'m','size':20}
font2 = {'family':'serif','color':'c','size':15}

plt.xlabel("Eje X", fontdict=font2)
plt.ylabel("Eje Y", fontdict=font2)
plt.title("Gráfica de la función x^2", fontdict=font1)
plt.grid(axis="y")#Sirve para poner la retícula sólo de los puntos del eje y
plt.show()
```

```python
x=np.linspace(0,20,20)
y1=x**2
plt.plot(x,y1, ls=":", linewidth=4)

font1 = {'family':'serif','color':'m','size':20}
font2 = {'family':'serif','color':'c','size':15}

plt.xlabel("Eje X", fontdict=font2)
plt.ylabel("Eje Y", fontdict=font2)
plt.title("Gráfica de la función x^2", fontdict=font1)
plt.grid(color="g", ls=":", linewidth=2)#Sirve para poner la retícula
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #El primer número corresponde al número de filas, el sgundo al número de columnas, y el tercero, la posición de la gáfica en la que se está
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


#plot 1:
x = np.linspace(0,10,100)
y1=x**2

plt.subplot(1, 5, 1)
plt.plot(x,y1, ls=":", color="c")

#plot 2:
x = np.linspace(0,10,100)
y2=np.sin(x)

plt.subplot(1, 5, 3)
plt.plot(x,y2,  ls="-", color="b")

#plot 3:
x = np.linspace(0,10,100)
y3=np.cos(x)

plt.subplot(1, 5 , 5)
plt.plot(x,y3,  ls="-.", color="k")

plt.show()
```
