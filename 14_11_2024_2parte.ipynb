```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
def get_data(N):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1    #Esta es notación científica, significa 1 por 10 a la menos uno

  for i in range(N):
    y[i]=3*x[i]**2+5*x[i]+-2+5*(eps*(-1)**i)

  return x,y
```

```python
def plot_data(x,y,params):
  z=params[0] + params[1]*x + params[2]*x**2

  plt.figure()
  plt.plot(x,y,"bo", label="Samples")
  plt.plot(x,z,"-m", label="Model")
  plt.title("Cuadratic Regression")
  plt.xlabel("X data")
  plt.ylabel("Y data")
  plt.grid(lw=2)
  plt.legend()
  plt.show()
```

```python
def cuadratic_regression(N=20):
   x,y=get_data(N)
   A=np.vstack([np.ones(len(x)),x, x**2]).T
   AtA=A.T@A
   Aty=A.T@y
   params=LA.solve(AtA,Aty)
   plot_data(x,y,params)
   OriData=np.array([-2.,5.,3.])
   error=np.round(LA.norm(params-OriData),3)
   print("error: ", error*100,"%")
   return params
```

```python
cuadratic_regression()
```
