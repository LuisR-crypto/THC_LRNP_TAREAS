```python
#Cargamos la base de datos de GitHub
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

link="https://github.com/LuisR-crypto/THC_LRNP_TAREAS/blob/main/EXECUTION_USA.csv"
code="?raw=true"
url=link+code
df=pd.read_csv(url, sep=",", index_col=1)
df.head()
```

```python
df.describe()
```

La descripción de la tabla, dice que tanto en la columna de la edad como en la de la cuenta de víctivas hay 1442 valores válidos, despues viene el pormedio de los datos, donde para la edad es de 41 años y medio, mientras que el número de víctimas era de una y media, después viene la variación estandar, que es que tanto se separan los valores del promedio aporximadamente, luego se muestra el valor mínimo que en la edad es de 22 y en las vícitmas es de 1.

Posterior llegan los porcentajes donde el 25% es el valor que separa el 25% de los datos más bajos, por lo que el 75% de las personas de la base de datos tienen edades mayores a 34 años y mayores o iguales a una víctima, lo mismo pasa en el de 50% y 75%, lo que sirve para ver en donde cae el mayor número de personas o entradas de la base de datos.

Por último, el valor máximo quese encuentra en cada una de las columnas, que es de 77 años y 168 víctimas respectivamente.


```python
#Graficamos el conteo de mujeres y hombres

sexo = df["Sex"].value_counts()

sexo.plot(kind = 'bar', color = ['blue','pink'])
plt.title('Distribucion por sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.xticks(rotation = 0)

plt.savefig("DistribucionSex.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Una gráfica de cantidad de blancos contra cantidad de negros y de latinos

raza = df["Race"].value_counts()

raza.plot(kind = 'bar', color = ['green'])
plt.title('Distribucion por raza')
plt.xlabel('Raza')
plt.ylabel('Cantidad')
plt.xticks(rotation = 0)

plt.savefig("DistribucionRace.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["Decada"]=(df["Date"].dt.year//10).astype(str)+"0"+"-"+(df["Date"].dt.year//10+1).astype(str)+"0"
decadas = df["Decada"].value_counts().sort_index()

decadas.plot(kind = 'bar', color = ['gray'])
plt.title('Distribucion de ejecuciones por década')
plt.xlabel('Decada')
plt.ylabel('Cantidad de ejecuciones')
plt.xticks(rotation = 0)
plt.savefig("DistribucionEjecuionesDecadas.png", dpi=300, bbox_inches="tight")
plt.show()


```

```python
#Una gráfica de la cantidad de ejecutados por estado
estados = df["State"].value_counts()

plt.figure(figsize=(10, 6))
estados.plot(kind = 'bar', color = ['purple'])
plt.title('Distribucion por estado')
plt.xlabel('Estado')
plt.ylabel('Cantidad')
plt.xticks(rotation = 0)
plt.savefig("DistribucionState.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Una gráfica de la cantida de personas ejecutadas por cada método
metodo = df["Method"].value_counts()

metodo.plot(kind = 'bar', color = ['darkred'])
plt.title('Distribucion por Método de ejecución')
plt.xlabel('Menores de edad')
plt.ylabel('Cantidad')
plt.xticks(rotation = 0)

plt.savefig("DistribucionMethod.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Cantidad de personas mayores contra personas menores
jovenes = df["Juvenile"].value_counts()

jovenes.plot(kind = 'bar', color = ['lightblue'])
plt.title('Distribucion por edad legal')
plt.xlabel('Menores de edad')
plt.ylabel('Cantidad')
plt.xticks(rotation = 0)

plt.savefig("DistribucionEdad.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
edad = df["Age"].value_counts().sort_index()

plt.figure(figsize=(15, 6))
edad.plot(kind = 'bar', color = ['darkgreen'])
plt.title('Distribucion por edad de los ejecutados')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.xticks(rotation = 0)

plt.savefig("DistribucionAge.png", dpi=300, bbox_inches="tight")
plt.show()
```
