```python
import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(-10,10,10000)
y1=np.sin(x**2)
plt.subplot(2,2,1)
plt.plot(x,y1, ls=":", color="r")

plt.xlim(-10,10)
plt.ylim(1.5, -1.5)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(["Función sin (x^2)"], loc="upper left", fontsize=5)
plt.grid()


x=np.linspace(-10,10,500)
y2=np.tanh(x)
plt.subplot(2,2,2)
plt.plot(x,y2, ls="-.", color="c")

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(["Función tanh(x)"], loc="upper left", fontsize=5)
plt.grid()


x=np.linspace(-10,10,500)
y3=np.sin(x)/x
plt.subplot(2,2,3)
plt.plot(x,y3, ls="-", color="m")

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(["Función tanh(x)"], loc="upper right", fontsize=5)
plt.grid()


x=np.linspace(-10,10,500)
y4=np.exp(-x/10)*np.sin(x)
plt.subplot(2,2,4)
plt.plot(x,y4, ls="--", color="y")

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(["Función exp(-x/10)"], loc="upper right", fontsize=5)
plt.grid()

plt.savefig("Subploat.png", dpi=300, bbox_inches="tight")
plt.show()
```
